# Matplotlib

### Overview

Matplotlib is the foundational plotting library in Python, known for its flexibility and extensive customization capabilities. It's widely used for creating static, publication-quality plots.

**Advantages**
- Flexibility: Highly customizable, allowing precise control over every element of the plot.
Wide Range of Plots: Supports a variety of plots including line, bar, scatter, histogram, and more.
- Integration: Integrates well with other libraries like NumPy and Pandas.
Publication Quality: Suitable for creating publication-quality figures.

**Best Use Cases** 
- Scientific and Academic Research: When detailed control over plot appearance is required.
- Publication-Ready Figures: For creating figures that need to meet strict formatting requirements.
- Static Plots: Ideal for static visualizations and when interactivity is not required.

#### Streaming and Real-Time Data

```python
# Basic Streaming
from bokeh.models import ColumnDataSource
from bokeh.plotting import curdoc, figure
from bokeh.driving import linear

source = ColumnDataSource(dict(x=[], y=[]))

p = figure()
p.line(source=source, x='x', y='y')

@linear()
def update(step):
    new_data = dict(x=[step], y=[step**0.5])
    source.stream(new_data, rollover=200)

curdoc().add_periodic_callback(update, 100)
show(p)
```

# Altair

### Overview

Altair is a declarative statistical visualization library for Python, based on Vega and Vega-Lite visualization grammars. It emphasizes simplicity, conciseness, and clarity, making it easy to create complex visualizations with minimal code.

**Advantages**
- Declarative Syntax: Uses a simple and human-readable syntax that abstracts away the complexities of the underlying plotting mechanisms.
- Ease of Use: Highly intuitive, making it easy for users to create visualizations with minimal code.
- Integration with Pandas: Works seamlessly with Pandas DataFrames, simplifying the process of visualizing data stored in these structures.
- Interactivity: Provides built-in interactivity for common visualizations, such as zooming, panning, and tooltips.
- Automatic Handling of Complex Tasks: Automatically handles common visualization tasks such as faceting, stacking, and binning, reducing the need for manual adjustments.

**Best Use Cases**
- Exploratory Data Analysis (EDA): Perfect for quickly generating a wide variety of visualizations to explore datasets.
- Statistical Visualization: Ideal for creating statistical visualizations that require detailed insights and easy customization.
- Interactive Reports: Suitable for creating interactive visualizations within reports and presentations, especially when integrated with Jupyter Notebooks.

```python
import altair as alt
import pandas as pd

# Sample data
data = pd.DataFrame({
    'sepal_length': [5.1, 4.9, 4.7, 4.6, 5.0, 5.4, 4.6],
    'sepal_width': [3.5, 3.0, 3.2, 3.1, 3.6, 3.9, 3.4],
    'species': ['setosa', 'setosa', 'setosa', 'setosa', 'setosa', 'setosa', 'setosa']
})

# Creating a scatter plot
scatter_plot = alt.Chart(data).mark_point().encode(
    x='sepal_length',
    y='sepal_width',
    color='species'
).properties(
    title='Scatter Plot of Sepal Length vs Sepal Width'
)

# Display the plot
scatter_plot.show()
```

### Cheat Sheet

#### Importing

```python
import altair as alt
import pandas as pd
```

#### Basic Plot Setup

```python
# Scatter Plot
data = pd.DataFrame({
    'x': [1, 2, 3, 4, 5],
    'y': [2, 3, 4, 5, 6]
})

scatter_plot = alt.Chart(data).mark_point().encode(
    x='x',
    y='y'
)
scatter_plot.show()

# Line Plot
line_plot = alt.Chart(data).mark_line().encode(
    x='x',
    y='y'
)
line_plot.show()

# Bar Plot
bar_plot = alt.Chart(data).mark_bar().encode(
    x='x',
    y='y'
)
bar_plot.show()
```

#### Advanced Plot Types

```python
# Histogram
histogram = alt.Chart(data).mark_bar().encode(
    alt.X('x', bin=True),
    y='count()'
)
histogram.show()

# Box Plot
box_plot = alt.Chart(data).mark_boxplot().encode(
    x='x',
    y='y'
)
box_plot.show()

# Heatmap
heatmap_data = pd.DataFrame({
    'x': [1, 1, 1, 2, 2, 2, 3, 3, 3],
    'y': [1, 2, 3, 1, 2, 3, 1, 2, 3],
    'value': [1, 2, 3, 4, 5, 6, 7, 8, 9]
})

heatmap = alt.Chart(heatmap_data).mark_rect().encode(
    x='x:O',
    y='y:O',
    color='value:Q'
)
heatmap.show()
```

#### Markdown

```python
# Titles and Labels
scatter_plot = scatter_plot.properties(
    title='Scatter Plot Title'
).configure_axis(
    titleFontSize=14,
    labelFontSize=12
)

# Axis Configuration
scatter_plot = scatter_plot.encode(
    alt.X('x', axis=alt.Axis(title='X-Axis Title')),
    alt.Y('y', axis=alt.Axis(title='Y-Axis Title'))
)

# Legends
scatter_plot = scatter_plot.encode(
    color=alt.Color('x', legend=alt.Legend(title='Legend Title'))
)
```

#### Interactivity

```python
# Adding Tooltips
scatter_plot = scatter_plot.encode(
    tooltip=['x', 'y']
)

# Adding Selection
selection = alt.selection_single(fields=['x'])

scatter_plot = scatter_plot.add_selection(
    selection
).encode(
    color=alt.condition(selection, 'x', alt.value('lightgray'))
)
```

#### Faceting

```python
# Row Facet
row_facet = scatter_plot.encode(
    alt.Row('y')
)
row_facet.show()

# Column Facet
column_facet = scatter_plot.encode(
    alt.Column('y')
)
column_facet.show()

# Facet Grid
facet_grid = scatter_plot.encode(
    alt.Facet('y', columns=2)
)
facet_grid.show()
```

#### Combining Charts

```python
# Layering 
line = alt.Chart(data).mark_line().encode(
    x='x',
    y='y'
)

points = alt.Chart(data).mark_point().encode(
    x='x',
    y='y'
)

layered_chart = alt.layer(line, points)
layered_chart.show()

# Concatenating Charts
chart1 = alt.Chart(data).mark_bar().encode(
    x='x',
    y='y'
)

chart2 = alt.Chart(data).mark_line().encode(
    x='x',
    y='y'
)

concat_chart = alt.concat(chart1, chart2)
concat_chart.show()
```

#### Example

```python
import altair as alt
import pandas as pd

# Sample Data
data = pd.DataFrame({
    'x': [1, 2, 3, 4, 5],
    'y': [2, 3, 4, 5, 6]
})

# Create a scatter plot
scatter_plot = alt.Chart(data).mark_point().encode(
    x='x',
    y='y',
    tooltip=['x', 'y']
).properties(
    title='Comprehensive Example Plot'
)

# Create a line plot
line_plot = alt.Chart(data).mark_line().encode(
    x='x',
    y='y'
)

# Layer scatter plot and line plot
layered_chart = alt.layer(scatter_plot, line_plot)

# Add customization for titles and labels
layered_chart = layered_chart.properties(
    title='Customized Comprehensive Example Plot'
).configure_axis(
    titleFontSize=14,
    labelFontSize=12
)

# Add hover tool
hover = alt.selection_single(fields=['x'], nearest=True, on='mouseover')

points = scatter_plot.mark_point().encode(
    opacity=alt.value(0)
).add_selection(
    hover
)

tooltips = scatter_plot.mark_text(align='left', dx=5, dy=-5).encode(
    text=alt.condition(hover, 'y', alt.value(' '))
)

final_chart = alt.layer(line_plot, points, tooltips)

# Show the plot
final_chart.show()
```
